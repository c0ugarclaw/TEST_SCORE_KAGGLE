```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('data/train.csv')

num_cols = ['age', 'study_hours', 'class_attendance', 'sleep_hours']
cat_cols = ['gender', 'course', 'internet_access',
            'sleep_quality', 'study_method',
            'facility_rating', 'exam_difficulty']

for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

ord_cols = ['gender', 'course', 'internet_access', 'sleep_quality', 'study_method', 'facility_rating', 'exam_difficulty']  # your categorical cols

enc = OrdinalEncoder()
df[ord_cols] = enc.fit_transform(df[ord_cols])
```

```python
output_df = df.pop('exam_score')
input_df = df.drop('id', axis=1)
x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size=0.2, random_state=42 )
```

sklearn Linear Regression

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(model.score(x_test, y_test))
```

```python
test_df = pd.read_csv('data/test.csv')
score_df = pd.read_csv('data/sample_submission.csv')
id = score_df['id']

num_cols = ['age', 'study_hours', 'class_attendance', 'sleep_hours']
cat_cols = ['gender', 'course', 'internet_access',
            'sleep_quality', 'study_method',
            'facility_rating', 'exam_difficulty']

for col in num_cols:
    test_df[col].fillna(test_df[col].mean(), inplace=True)

for col in cat_cols:
    test_df[col].fillna(test_df[col].mode()[0], inplace=True)


ord_cols = ['gender', 'course', 'internet_access', 'sleep_quality', 'study_method', 'facility_rating', 'exam_difficulty']  # your categorical cols

enc = OrdinalEncoder()
test_df[ord_cols] = enc.fit_transform(test_df[ord_cols])

test_x = test_df.drop('id', axis=1)

test_pred = model.predict(test_x)
submission = pd.DataFrame({
    "id": id,
    "exam_score": test_pred
})

submission.to_csv("data/test_sk.csv", index=False)
```
